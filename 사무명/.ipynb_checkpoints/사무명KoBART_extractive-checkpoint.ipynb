# Tokenization

```python
!pip install transformers
import torch
from transformers import PreTrainedTokenizerFast
from transformers import BartForConditionalGeneration

tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization')
```

```python
# 'Text' 열의 데이터 tokenization
text_tokens = dataset['Text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, max_length=512))

# 'Summary' 열의 데이터 tokenization
summary_tokens = dataset['Summary'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, max_length=128))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization')
model.to(device)
```

```python
BATCH_SIZE = 1
```

```python
from torch.utils.data import DataLoader, Dataset

class SummaryDataset(Dataset):
    def __init__(self, texts, summaries):
        self.texts = texts
        self.summaries = summaries

    def __getitem__(self, idx):
        return {
            'input_ids': torch.tensor(self.texts.iloc[idx]),
            'labels': torch.tensor(self.summaries.iloc[idx])
        }

    def __len__(self):
        return len(self.texts)

# Train-test split 이후의 데이터를 바탕으로 Dataset을 생성
train_dataset = SummaryDataset(text_tokens.loc[dataset_train.index], summary_tokens.loc[dataset_train.index])
test_dataset = SummaryDataset(text_tokens.loc[dataset_test.index], summary_tokens.loc[dataset_test.index])

train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=BATCH_SIZE)
test_dataloader = DataLoader(test_dataset, shuffle=False, batch_size=BATCH_SIZE)
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    input_ids = [item['input_ids'] for item in batch]
    labels = [item['labels'] for item in batch]

    # pad_sequence를 사용하여 동일한 길이로 패딩합니다.
    input_ids = pad_sequence(input_ids, batch_first=True, padding_value=tokenizer.pad_token_id)
    labels = pad_sequence(labels, batch_first=True, padding_value=tokenizer.pad_token_id)

    return {
        'input_ids': input_ids,
        'labels': labels
    }

train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=BATCH_SIZE, collate_fn=collate_fn)
test_dataloader = DataLoader(test_dataset, shuffle=False, batch_size=BATCH_SIZE, collate_fn=collate_fn)
```

```python
from transformers import AdamW, get_linear_schedule_with_warmup

model = BartForConditionalGeneration.from_pretrained('digit82/kobart-summarization').to(device)

optimizer = AdamW(model.parameters(), lr=5e-5)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=len(train_dataloader) * 3)
```

```python
epochs = 5
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for idx, batch in enumerate(train_dataloader):
        optimizer.zero_grad()

        inputs = batch['input_ids'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids=inputs, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        optimizer.step()
        scheduler.step()

    avg_train_loss = total_loss / len(train_dataloader)
    print(f"Epoch {epoch+1}/{epochs} - Training loss: {avg_train_loss:.4f}")
```

```python
model.eval()  # 평가 모드로 모델 설정

generated_summaries = []

# 테스트 데이터셋을 통한 요약 생성
for batch in test_dataloader:
    with torch.no_grad():  # 기울기 계산 비활성화
        input_ids = batch['input_ids'].to(device)
        summary_ids = model.generate(input_ids, max_new_tokens=100)

        # 생성된 요약 디코딩 및 저장
        for s in summary_ids:
            generated_summaries.append(tokenizer.decode(s, skip_special_tokens=True))

# 생성된 요약과 실제 요약 비교
true_summaries = dataset_test['Summary'].tolist()

# 예: ROUGE 지표를 사용하여 성능 측정
# (ROUGE 계산을 위한 추가 라이브러리 설치 및 구현이 필요합니다.)
# rouge_score = compute_rouge(generated_summaries, true_summaries)
# print(rouge_score)
```

```python
from rouge import Rouge

# 모델을 평가 모드로 설정
model.eval()

# 생성된 요약과 실제 요약을 저장할 리스트 초기화
generated_summaries = []
real_summaries = []

# 테스트 데이터셋을 순회하면서 요약 생성
for idx, row in dataset_test.iterrows():
    input_text = row['Text']
    input_ids = tokenizer.encode(input_text, add_special_tokens=True, truncation=True, max_length=512, return_tensors="pt").to(device)
    with torch.no_grad():
        summary_ids = model.generate(input_ids, max_length=103)
    generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    generated_summaries.append(generated_summary)
    real_summaries.append(row['Summary'])

# ROUGE 점수 계산
rouge = Rouge()
scores = rouge.get_scores(generated_summaries, real_summaries, avg=True)

print(scores)
```

```python
import matplotlib.pyplot as plt

# ROUGE 스코어 데이터
scores = {
    'rouge-1': {'r': 0.4848754668270797, 'p': 0.48921967743174793, 'f': 0.4791588663865904},
    'rouge-2': {'r': 0.3910931174089069, 'p': 0.3796176046176046, 'f': 0.38482702218409964},
    'rouge-l': {'r': 0.4800582625260044, 'p': 0.4751456033576738, 'f': 0.47198526523881423}
}

labels = list(scores.keys())
recall = [score['r'] for score in scores.values()]
precision = [score['p'] for score in scores.values()]
f_score = [score['f'] for score in scores.values()]

# 히스토그램 생성
bar_width = 0.25
r1 = range(len(recall))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# 색상 설정
colors = ["#132C33", '#17869E', '#B4DBE9']  # 연한 노랑, 연한 초록, 연한 핑크

plt.bar(r1, recall, color=colors[0], width=bar_width, edgecolor='grey', label='Recall')
plt.bar(r2, precision, color=colors[1], width=bar_width, edgecolor='grey', label='Precision')
plt.bar(r3, f_score, color=colors[2], width=bar_width, edgecolor='grey', label='F-score')

# 제목 및 레이블 추가
plt.title('ROUGE scores')
plt.xlabel('Metrics', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(recall))], labels)
plt.legend()

plt.show()
```
